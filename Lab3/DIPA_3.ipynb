```python
gray_2= gray[:,:gray.shape[0]-1] #Squaring the image
```

```python
cv2_imshow(gray_2)
```

```python
pool_size = 5
new_img =[]
temp =[]

def sum1(input):
    return sum(map(sum, input))

for i in range(0,gray_2.shape[0],pool_size):
  temp=[]
  for j in range(0,gray_2.shape[1],pool_size):
      temp.append(sum1(gray_2[i:i+pool_size,j:j+pool_size])/(pool_size**2))
  new_img.append(temp)
```

```python
cv2_imshow(np.array(new_img))
```
