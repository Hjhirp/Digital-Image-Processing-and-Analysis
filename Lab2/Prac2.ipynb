```python
import skimage.io
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import math
```

```python
def LogTransform(image):
  maxValue = 255
  c=255/(math.log(1+maxValue))
  new_img=[]
  for i in range(image.shape[0]):
    temp=[]
    for j in range(image.shape[1]):
      temp.append(c*math.log(image[i][j]+1))
    new_img.append(temp)
  return np.array(new_img)
```

```python
def GammaTransform(image,gamma):

  new_img=[]
  for i in range(image.shape[0]):
    temp=[]
    for j in range(image.shape[1]):
      temp.append(255*(image[i][j]/255)**gamma)
    new_img.append(temp)
  return np.array(new_img)
```

```python
def ContrastStretching(image):
  maxiI = np.amax(image) - 5
  miniI = np.amin(image) + 5

  maxoI = np.amax(image)*2/3
  minoI = np.amax(image)*1/3

  new_img = image.copy()
  for i in range(image.shape[0]):
    for j in range(image.shape[1]):
      pixel = new_img[i, j]
      pixel = (pixel - miniI) * ((maxoI-minoI) / (maxiI-miniI)) + minoI
      new_img[i, j] = pixel
  return np.array(new_img)
```

```python
import glob

list_ = glob.glob("/content/drive/MyDrive/DIPA/Images/*.jpg")
gamma=float(input("Enter gamma: "))
for name in list_[0:3]:
  image = skimage.io.imread(name)
  print('Original')
  cv2_imshow(image)

  print('Log Transform')
  new_img = LogTransform(image)
  cv2_imshow(new_img)

  print('Gamma Transform')
  new_img = GammaTransform(image,gamma)
  cv2_imshow(new_img)

  print('Contrast Stretching')
  new_img = ContrastStretching(image)
  cv2_imshow(new_img)
```
