Algorithm to enlarge or shrink -->

*   capture shape of image in two variable
*   **new size -- Old size * factor || same for y**
*   **xScale -- xNew/(w-1) || same for y**
*   create a new image with shape of new size
*   for loops to iterate the image
*   new image[i+1,j+1] = old_image[1+ int(i/xscale), 1 + int(j/yScale)]
*   voila we have the new image.




```python
w, h = gray1.shape[:2];

size= float(input('Enter the size factor: '))
xNew = int(w * size);
yNew = int(h * size);

xScale = xNew/(w-1);
yScale = yNew/(h-1);

newImage = np.zeros([xNew, yNew, 3]);

for i in range(xNew-1):
   for j in range(yNew-1):
       newImage[i + 1, j + 1]= gray1[1 + int(i / xScale), 1 + int(j / yScale)]

cv2.imwrite('shrinked_image.jpg',newImage)

cv2_imshow(newImage)
```
